## Recursive Function 

```python
"""
--- When function call itself again anf again then it is called recursive 
function 
        Syntax:
        
        def func():
          Statement1
          Statement2
          Statement3
          Statement4
          func()
          
        func()


"""
```

```python
def count_down(start):
  print(start)

  if start>1:
    count_down(start-1)
```

```python
count_down(10)
```

## Lambda Function 

```python
"""
--- It is an anonymous function , because it dont have name.
--- This fucntion is define using lambda keyword so it is called as lambda
function.
--- This function use only once or short period of time.
    
        Syntax:
            lambda arguments: expression 
        arguments ----> no of argumentswe can pass 
        expression -----> It will be only one 

"""
```

```python
add = lambda n,m : n+m
```

```python
add(1,2)
```

```python
greater = lambda m,n: m>n if m>n else print(f"{m} is Greater than {n}")
greater(42)
```

```python
greater = lambda m,n: print(f"{m} is greater than {n}") if m>n else print(f"{n} is greater than {m}") 
greater(20,10)
```

```python
list1 = [1,2,3,4,5]
sqr = lambda m : [i**2 for i in list1]  
sqr(list1)
```

```python
ss = lambda a : a**2
```

```python
for i in list1:
  print(ss(i))
```

## FILTER FUNCTION IN PYTHON

```python
"""
FILTER FUNCTION IN PYTHON

--- THE FILTER() FUNCTION IN PYTHON IS A BUILD-IN FUCNTION THAT ALLOWS YOU TO FILTER AN ITERABLE (LIST,TUPLE,STR) BASED ON GIVEN CONDITION 
--- IT RETURN AN ITERATOR CONTAINING AN ELEMENT FROM THE ITERABLE FOR WHICH THE FUNCTION RETURNS TRUE
        SYNTAX:
          filter(function,iterable)


"""
```

```python
def even_number(num):
    if num%2==0:
        return True
```

```python
list6 = [1,2,3,4,5,6,7,8,10]

list(filter(even_number,list6))
```

```python
even_number = lambda n : n%2==0 
c = list(filter(even_number,[1,2,3,4,5,6,7,8,9]))
c
```
