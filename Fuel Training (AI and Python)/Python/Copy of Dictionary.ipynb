```python
lis1 = [2,1,2,4,5,4,6,9,9,2]
list_unique=[]
list_duplicate=[]
for i in lis1:
  if i not in list_unique:
    list_unique.append(i)
  else:
    if i not in list_duplicate:
        list_duplicate.append(i)

print(list_unique)
print(list_duplicate)
```

```python
lis1 = [1,2,3,4,5,6,1,2,3,4,5,6,1,2,3,4,5,6]
list_duplicate=list(set(x for x in lis1 if lis1.count(x)>1))
list_unique = list(set([i for i in lis1]))
# list_unique =[i for i in lis1 if i list_duplicate]
print(list_duplicate)
print(list_unique)
```

```python
lis5 = [1,2,3,4,5,6,1,2,3,4,5,6,1,2,3,4,5,6]
uni_list = list(set([i for i in lis5]))
dupli_list = [i for i in uni_list if lis5.count(i)>1]
print(uni_list)
print(dupli_list)
```

```python
dict5 = {"Name":"Rohit","Age":35,"Type of Player":"Batsman","ODI ranking":2}
myKeys = list(dict5.keys())
myKeys.sort()
sorted_dict = {i: dict5[i] for i in myKeys}
 
print(sorted_dict)

```

```python
marks = {"Maths":97,"Phy":95,"Chem":90,"Bio":94}
sorted_marks = {}
for i in sorted(marks):
  sorted_marks[i] = marks[i]
print(sorted_marks)
```
