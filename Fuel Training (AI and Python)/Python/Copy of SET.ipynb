```python
"""
1.Union 
-- The Union operations on set return the set that contain all unique element from both set
-- In other words , it creates a new set that contain all element from both set without duplicate
-- we can pass number of sets.

        SYNTAX:
              set1.union(set1,set2,set3,........)


"""
```

```python
set1 = {1,2,3}

set2 = {3,5,6,8}

set3 = set1.union(set2)
set3
```
